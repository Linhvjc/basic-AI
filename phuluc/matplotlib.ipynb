### Biu  ng t gif file

```python
import numpy as np
import matplotlib.pyplot as plt
import imageio
import os

def animate(i):
    x = np.linspace(0, 2, 1000)
    y = np.sin(2 * np.pi * (x - 0.01 * i))
    return x, y

def plot(i):
  x, y = animate(i)
  fig = plt.figure(figsize=(8, 6))
  plt.plot(x, y)
  plt.xlabel('x')
  plt.ylabel('y')
  plt.title("iteration {}".format(i))
  fig.savefig("gifplot/{:03d}.png".format(i))
  plt.close(fig)
  
for i in range(20):
    plot(i)

with imageio.get_writer('mygif.gif', mode='I') as writer:
    for filename in os.listdir('gifplot'):
        image = imageio.imread('gifplot/{}'.format(filename))
        writer.append_data(image)

from IPython.display import Image
Image(open('mygif.gif','rb').read())
```

```python
# get mean of passenger in each year
df_summary = flights[['year', 'passengers']].groupby('year').mean()
x = list(df_summary.index)
y = [int(people) for people in df_summary['passengers']]
x, y
```

```python
plt.figure(figsize=(16,8))
plt.plot(x, y, marker = 'o', linewidth = 2)
plt.xlabel('Year')
plt.ylabel('passengers')
plt.title('Passengers in each year')
```
