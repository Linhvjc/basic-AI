### 1.2 Véc tơ

```python
x = torch.tensor([1, 1.2, 1.5, 1.8, 2])
```

```python
# Độ dài
print("length of vector: ", x.size()) # or len(x)

# Định dạng của véc tơ
print("vector type: ", x.dtype)

# Tổng của các phần tử 
print("sum of vector: ", x.sum())

# Trung bình các phần tử
print("mean of vector: ", x.mean())

# Giá trị nhỏ nhất
print("min of vector: ", x.min())

# Giá trị lớn nhất
print("max of vector: ", x.max())
```

```python
x = torch.tensor([1, 2, 1.5, 1.8, 1.9])
y = torch.tensor([1.1, 2.2, 1.2, 1.6, 1.7])
print("x + y: ", x + y)
print("x - y: ", x - y)
print("x * y: ", x * y)
```

```python
x = torch.tensor([1, 2, 1.5, 1.8, 1.9])
print("x + 5: ", x + 5)
print("x - 5: ", x - 5)
print("x * 5: ", x * 5)
```

### 1.3 Ma trận

```python
A = torch.tensor([[1, 2, 3],
                  [3 , 2 , 1]])
# Shape cua matrix A
A.size()
```

```python
import torch
A = torch.tensor([[1, 2, 3], 
                  [3, 2, 1]])

B = torch.tensor([[2, 1, 2], 
                  [1, 3, 0]])

A*B
```

```python
print("A-B: \n", A-B)
print("A+B: \n", A+B)
```

```python
import torch
A = torch.tensor([[1, 2, 3], 
                  [3, 2, 1]])

B = torch.tensor([[2, 1], 
                  [1, 3],
                  [1, 1]])

A@B
```

```python
# Tim ma tra nghich dao
A = torch.tensor([[1, 2, 3],
                  [3, 2, 1],
                  [4, 2, 2]], dtype=torch.float)
A_inv = torch.linalg.pinv(A)
A_inv
```

```python
A = torch.tensor([[1, 2, 3],
                  [3, 2, 1],
                  [4, 2, 2]])
# truy cap ma tran con tu 2 dong dau tien va 2 cot dau tien
A[:2, :2]
```

```python
# Truy cap mot index cu the, vi du dong 0, 2 cua ma tran
A.index_select(0, torch.tensor([0,2]))
```

### 1.4 Tensor

```python
A = torch.tensor([[[1, 2, 3], 
                  [3, 2, 1]],
                  [[2, 1, 2], 
                  [1, 3, 0]]])

# Kích thước của tensor
print("shape of A: " , A.size())

# Số chiều 
print("total dim: ", A.ndim)

# Định dạng dữ liệu
print("dtype: ", A.dtype)
```

```python
import torch

A = torch.randn([2, 3, 4])
B = torch.randn([2, 4, 2])

# Tích giữa 2 tensor
(A@B).size()
```

```python
A = torch.randn([2, 3, 4])
B = torch.randn([2, 3 , 4])
# Tichs hadamard giua 2 tensor
(A*B).size()
```

```python
# Phep cong
(A+B).size(), (A-B).size()
```

```python
import torch

# Khởi tạo ma trận A kích thước m, n, p = 2, 3, 4
A = torch.randn([2, 3, 4])

# Truy cập ma trận đầu tiên 
A[:1, :, :]
```

```python
# Truy cập tương ứng với các chiều m, n, p lần lượt index đầu tiên, 2 index đầu tiên, và index thứ 3.
A[:1, :2, 3]
```

### 1.5 Tích giữa một ma trận với vecto 

```python
X = torch.tensor([[100, 120, 80, 90, 105, 95], 
                  [2, 3, 2, 2, 3, 2]])
```

```python
w = torch.tensor([[10], [100]])
y = X.T@w       #X.T là ma trận chuyển vị
y
```
